# **🛠️ 4. Pré-Processamento de Dados**

O pré-processamento é uma etapa crucial para preparar os dados para a modelagem. Aqui, garantimos que a máquina receba as informações no formato correto. Nesta etapa, realizaremos a **separação dos dados**, dividindo nosso dataset entre variáveis preditivas (X) e a variável alvo (y), e posteriormente separando-os em conjuntos de treino e teste.

```python
# 1. Separando as features (X) e o alvo (y)
# No seu projeto, as features são as contagens de Homens e Mulheres
X = df_final[['Homens', 'Mulheres']]
y = df_final['Total'] # O Total é o que queremos "prever" ou analisar

print("✅ Variáveis X e y separadas com sucesso!")
```

```python
from sklearn.model_selection import train_test_split

# Dividindo os dados: 70% para treino e 30% para teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Exibindo as dimensões (o tamanho das fatias)
print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")
```

### **4.1. Padronização (StandardScaler)**

A padronização é uma técnica que reajusta os dados para que tenham média 0 e desvio padrão 1. Diferente da normalização, ela não limita os dados a um intervalo fixo, sendo ideal para algoritmos que assumem que os dados seguem uma distribuição normal.

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 1. Inicializar o StandardScaler
scaler_std = StandardScaler()

# 2. Ajustar e transformar os dados de treino
X_train_std = scaler_std.fit_transform(X_train)

# 3. Criar o DataFrame com o nome correto para evitar erros
df_standardized = pd.DataFrame(X_train_std, columns=X_train.columns)

# 4. Transformar os dados de teste (sem o fit)
X_test_std = scaler_std.transform(X_test)

print("✅ Padronização concluída com sucesso!")
```

```python
# Exibir as primeiras 5 linhas dos dados padronizados (treino)
print("\nPrimeiras 5 linhas dos dados padronizados (treino):")
df_standardized.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

# Criando o histograma da coluna 'Mulheres' padronizada
sns.histplot(df_standardized['Mulheres'], kde=True, color='indianred')

plt.title('Distribuição da População Feminina (Padronizada)')
plt.xlabel('Desvios Padrão em relação à Média')
plt.ylabel('Frequência')

plt.show()
```

**Análise:** Com a padronização, os dados agora estão centralizados no zero. Isso permite comparar as variáveis de Homens e Mulheres em uma mesma escala estatística, facilitando a identificação de padrões demográficos sem a interferência da magnitude dos números originais do IBGE.
